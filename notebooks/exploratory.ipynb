```
import pandas as pd
from pathlib import Path
data = Path('data') / 'real_sample_orders.csv'
df = pd.read_csv(data, parse_dates=['timestamp'])
df.head()
```

```
import plotly.express as px
agg = df.groupby(df['timestamp'].dt.hour).order_id.count().reset_index()
agg.columns=['hour','orders']
fig = px.bar(agg, x='hour', y='orders', title='Orders by hour')
fig.show()
```

```
import matplotlib.pyplot as plt
plt.hist(df['processing_time_min'], bins=20)
plt.title('Processing time distribution')
plt.show()
```
